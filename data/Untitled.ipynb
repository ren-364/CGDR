```python
from collections import defaultdict
def fenlei(voc):
# 创建一个默认字典用于存储分组结果
    grouped_data = defaultdict(list)
# 按前三位字符进行分组
    for key, value in voc.word2idx.items():
        prefix = key[:3]
        grouped_data[prefix].append((key, value))
# 计算每个组的元素个数并输出结果
    result = {k: len(v) for k, v in grouped_data.items()}
    con_ = 0
# 打印结果
    for prefix, count in result.items():
        print(f"前三位字符 '{prefix}' 的组包含 {count} 个元素。")
        con_ = con_ + count
    print(f"共分为 '{len(result)}' 个类，共{con_}个元素")
```
